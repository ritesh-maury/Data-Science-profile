# Subplotting

```python

fig, axs = plt.subplots(2,2,figsize=(13,15))

df["Sex"].value_counts().plot(kind="bar",color=["red","green"],ax=axs[0][1])
axs[0][1].set_title("Number of Passengers with Sex columns")
axs[0][1].grid(True)

sns.countplot("Sex",hue="Survived",data=df,ax=axs[0][0])
axs[0][0].set_title("Survived with Gender columns")
axs[0][0].grid(True)

axs[1][0].scatter(df["Age"],df["Fare"])
axs[1][0].set_title("Scatter plotting between Age & Fare")

sns.boxplot(x="Survived",y="Age",data=df,ax=axs[1][1])
axs[1][1].set_title("Boxplot with Survived and Age columns")

plt.show()
```
